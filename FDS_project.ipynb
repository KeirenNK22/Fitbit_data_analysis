```python
fig, ax = plt.subplots(figsize=(8,5))

cols = ["TotalDistance", "TotalSteps", "SedentaryMinutes", "TotalActiveMinutes", "Calories"]
sns.heatmap(daily_activity[cols].corr(), cmap='RdBu', annot=True, annot_kws={'size': 15})

plt.title("Correlation Matrix")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))

cols=["Calories", "Intensity", "METs", "Steps"]
sns.heatmap(minute[cols].corr(), cmap="icefire", annot=True, annot_kws={'size': 15})

plt.title("Correlation Matrix")
plt.show()
```

```python
print(daily_activity["TotalActiveMinutes"].sum() / daily_activity['SedentaryMinutes'].sum())
```

```python
fig, ax = plt.subplots(figsize=(7,7))

x = np.array([daily_activity["SedentaryMinutes"].mean(), daily_activity["TotalActiveMinutes"].mean()])

plt.pie(x, labels=["Sedentary Minutes", "Total Active Minutes"], autopct='%1.0f%%') 
plt.title("Sedentary Minutes Vs Total Active Minutes")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,7))

plt.bar(daily_activity.Weekday, daily_activity.VeryActiveMinutes)

plt.xlabel("Day of the Week")
plt.ylabel("Number of Very Active Minutes")
plt.title("Very Active Minutes by Day of the Week")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,7))

plt.bar(daily_activity.Weekday, daily_activity.TotalSteps, color='#5e7fcc')

plt.xlabel("Day of the Week")
plt.ylabel("Average Steps")
plt.title("Average Number of Steps by Taken Day of the Week")
plt.show()
```

```python
sns.relplot(x="Calories", y="TotalSteps", data=daily_activity, kind="scatter")

plt.xlabel("Calories")
plt.ylabel("Total Steps")
plt.title("Relationship Between Calories and Total Steps")
plt.show()
```

```python
sns.relplot(x="Calories", y="SedentaryMinutes", data=daily_activity, kind="scatter")

plt.xlabel("Calories")
plt.ylabel("Sedentary Minutes")
plt.title("Relationship between Calories and Sedentary Minutes")
plt.show()
```

```python
sns.relplot(x="Calories", y="TotalIntensity", data=hourly, kind="scatter")

plt.title("Relationship between Calories and Total Intensity")
plt.xlabel("Calories")
plt.ylabel("Total Intensity")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,7))

hourly.groupby(hourly["hour"])["StepTotal"].mean().plot(kind='bar', rot=0, color='#8484cf')

plt.title("The Most and Least Active Hours of The Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Steps Taken in a Single Hour")
plt.show()
```
